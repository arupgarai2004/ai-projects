
## Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring and Assigning Variables

age = 30  # This is an integer variable
height = 5.9  # This is a float variable
name = "Arup"  # This is a string variable
is_student = True  # This is a boolean variable
## Printing the variables
print("Age:", age);
print("Height:", height);
print("Name:", name);
print("Is Student:", is_student);
```

```python
## Naming Conventions
# Variable names should be descriptive and follow certain conventions:
# They should start with a letter (a-z, A-Z) or an underscore (_). and  contain only letters, numbers, and underscores.
# Name should be case sensitive (e.g., age and Age are different variables).

#valid variable names
first_name = "Arup"
last_name = "Garai"

```

```python
## Understanding Variables and Data Types
## Python is dynamically typed, which means you dont need to declare the type of a variable when you create it. The type is inferred from the value assigned to the variable. You can also change the type of a variable by assigning a different type of value to it.
age=25 #int
height=6.1 #float
name="Arup" #str
is_student=True #bool
print(type(name))
```

```python
## type checking and type conversion
type(age) # Output: float

```

```python
age = 25
print(type(age)) # Output: <class 'int'>

# Type conversion
age_str = str(age)  # Convert integer to string
print(age_str)  # Output: '25'
print(type(age_str))  # Output: <class 'str'>
```

```python
age='25'
print(type(int(age)))
```

```python
## Dynamic Typing
# In Python, you can reassign a variable to a different type of value without any issues
var=10 #int
print(var,type(var))

var="Hello"
print(var,type(var))

var=3.14
print(var,type(var))
```
