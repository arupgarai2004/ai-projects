```python
## Arithmethic Operation

a = 10 
b = 5

addition = a + b ## Addition
print("Addition:", addition);
subtraction = a - b ## Subtraction
print("Subtraction:", subtraction);
multiplication = a * b ## Multiplication
print("Multiplication:", multiplication);
division = a / b ## Division
print("Division:", division);
modulus = a % b ## Modulus (Remainder)
print("Modulus:", modulus);
exponentiation = a ** b ## Exponentiation (a raised to the power of b)
print("Exponentiation:", exponentiation);
```

```python
## Comparison Operators
## == Equal to
a = 10
b = 10

print(a == b)  # Output: True
```

```python
## logical operators AND, OR, NOT
x = True
y = False     
print(x and y)  # Output: False
print(x or y)   # Output: True
print(not x)    # Output: False  
```

```python
## Simple Calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

## Perform arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "Undefined (division by zero)"
floor_division = num1 // num2 if num2 != 0 else "Undefined (division by zero)"
modulus = num1 % num2 if num2 != 0 else "Undefined (division by zero)"
exponentiation = num1 ** num2

## Display results
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```
