Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a language. In simpler terms, syntax is about the correct arrangement of words and symbols in a code.

Semantics refers to the meaning or the interpretation of the symbols, characters, and commands in a language. It is about what the code is supposed to do when it runs.

```python
## Basic Python Syntax Rules
### 1. Case Sensitivity - Python case sensitive

name = "Arup"
Name = "Garai"
print(name)  # Output: Arup
print(Name)  # Output: Garai
```

### Indentation
Indentation in Python is used to define the structure and hierarchy of the code. Unlike many other programming languages that use braces {} to delimit blocks of code, Python uses indentation to determine the grouping of statements. This means that all the statements within a block must be indented at the same level.

```python
## Indentation
## Python uses indentation to define blocks of code. Consistent use of spaces or tabs is crucial.
age = 25
Name = "Arup"
if age >= 18:
    print("You are an adult.");
    Name = "Garai"
else:
    print("You are a minor.");


print("This is outside the if-else block." + Name)  # Output: This is outside the if-else block. Garai
```

```python
## Line Continuation
## You can use a backslash (\) to indicate that a statement continues on the next line.
total = 1 + 2 + 3 + \
        4 + 5
print(total)  # Output: 15
```

```python
## Multiple Statements on a Single Line
## You can use a semicolon (;) to separate multiple statements on a single line, but it's generally not recommended for readability.
x = 10; y = 20; z = x + y
print(z)  # Output: 30
```

```python
## Understand semantics in python
# varibales assignments
age = 30  ## age is integer
name = "Arup"  ## name is string
is_student = True  ## is_student is boolean
print(type(age), type(name), type(is_student))  # Output: <class 'int'> <class 'str'> <class 'bool'>
```

```python
## Type Inference
# Python can infer the type of a variable based on the value assigned to it.
var1 = 10  # var1 is inferred as an integer
print(type(var1))  # Output: <class 'int'>
var1 = "ten"  # var1 is now inferred as a string
print(type(var1))  # Output: <class 'str'>
```
