## DataTypes
# 1. Definition:
Data types are a classification of data which tell the compiler or interpreter how the programmer intends to use the data.
They determine the type of operations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data.
# 2. Importance of Data Types in Programming
Explanation:

- Data types ensure that data is stored in an efficient way.
- They help in performing correct operations on data.
- Proper use of data types can prevent errors and bugs in the program.

```python
## Integer Example
age = 25
print(age)
print(type(age))  # Output: <class 'int'>

## Float Example
height = 6.1
print(height)
print(type(height))  # Output: <class 'float'>

## String Example
name = "Arup"
print(name)
print(type(name))  # Output: <class 'str'>

## Boolean Example
is_student = True
print(is_student)
print(type(is_student))  # Output: <class 'bool'>

a= 20   
b = 20
type(a==b) # Output: <class 'bool'>
```

```python
## common errors
result = "Hello" + 5  # This will raise a TypeError because you cannot concatenate a string and an integer.
```

```python
## correct way to concatenate string and integer
result = "Hello " + str(5)  # Convert the integer to a string before concatenation
print(result)  # Output: Hello 5
```

```python
## Advanced Data Types
## Lists like a JSON array
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits)
print(type(fruits))  # Output: <class 'list'>

## Tuples like a JSON array but immutable
coordinates = (10.0, 20.0)
print(coordinates)
print(type(coordinates))  # Output: <class 'tuple'>

## Dictionaries like a JSON object
person = {"name": "Arup", "age": 25, "is_student": True}
print(person)
print(type(person))  # Output: <class 'dict'>
```
